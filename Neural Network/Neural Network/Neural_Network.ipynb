```python
scaler = MinMaxScaler()
print(df)
```

```python
df[["BMI","GenHlth","MentHlth","PhysHlth","Age","Education","Income"]] = scaler.fit_transform(df[["BMI","GenHlth","MentHlth","PhysHlth","Age","Education","Income"]])
print(df)
```

```python
X = np.array(df[['HighBP', 'HighChol', 'CholCheck', 'BMI',
       'Smoker', 'Stroke', 'Diabetes', 'PhysActivity', 'Veggies',
       'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth',
       'MentHlth', 'PhysHlth', 'DiffWalk', 'Sex', 'Age','Fruits', 'Education','Income']])#, 'Fruits', 'Education','Income'
y = np.array(df['HeartDiseaseorAttack']).reshape(-1,1)
```

```python
polyfeature = PolynomialFeatures(degree=2)
X = polyfeature.fit_transform(X)
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=42)
X_train.shape
```

```python
diseasecase = np.sum(y==0)
print(diseasecase)
diseasecase = np.sum(y==1)
print(diseasecase)
```

```python
print(X.shape)
print(y.shape)
```

```python
model = Sequential([
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(8, activation="relu"),
    # Dense(4, activation="relu"),
    Dense(1, activation="linear")
])
```

```python
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,loss = BinaryCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
loss, acc = model.evaluate(X_train,y_train)
print(f"accuray of model is {acc*100}")
# relu relu softmax
loss, acc = model.evaluate(X_test,y_test)
print(f"accuray of model is {acc*100}")
```

```python
def recallex(y_pre,y_true):
  num = len(y_true)
  FP = 0
  FN = 0
  TP = 0
  TN = 0
  for i in range(num):
    if y_pre[i] == 0 and y_true[i]== 0:
      TN +=1
    if y_pre[i] == 0 and y_true[i]== 1:
      FN +=1
    if y_pre[i] == 1 and y_true[i]== 1:
      TP +=1
    elif y_pre[i] == 1 and y_true[i]== 0:
      FP +=1

  print(f' TN : {TN} \n FN : {FN} \n TP : {TP} \n FP : {FP}    ')

def result(X_train):
  predictions = model.predict(X_train)
  custom_threshold = 0.5

# Apply custom threshold to sigmoid output
  predict = (predictions >= custom_threshold).astype(int)
  return predict
```

```python
y_pred = result(X_train)
recallex(y_pred,y_train)
```

```python
y_pred = result(X_test)
recallex(y_pred,y_test)
```
