```python
class MyLinearRegression:
  def __init__(self,X,y):
    self.X_train = None
    self.X_test = None
    self.y_train = None
    self.y_test = None
    self.X = X
    self.y = y

  def data_split(self):
    self.X_train,self.X_test,self.y_train,self.y_test = train_test_split(self.X,self.y,test_size = 0.2,random_state = 42)

  def linearfitmodel(self):
    self.model = LinearRegression()
    self.model.fit(self.X_train,self.y_train)

  def logisticfitmodel(self):
    self.model = LogisticRegression()
    self.model.fit(self.X_train,self.y_train)

  def predict(self):
    self.y_predict = self.model.predict(self.X_test)
    mse = mean_squared_error(self.y_test,self.y_predict)
    print(f"Mean Square Error is : {mse}")

  def visual(self):
    plt.scatter(X,y,s = 5)
    plt.plot(self.X_test,self.y_predict,color = "red")


def myDataFrame():
  filepath = str(input("Enter your file path : "))
  df = pd.read_csv(filepath)
  X = df[["Years of Experience"]]
  y = df["Salary"]
  return X,y

# X,y = myDataFrame()
# mymodelforfit = MyLinearRegression(X,y)
# mymodelforfit.data_split()
# mymodelforfit.linearfitmodel()
# mymodelforfit.predict()
# mymodelforfit.visual()
```
