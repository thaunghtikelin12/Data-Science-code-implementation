```python


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBRFClassifier
from sklearn.metrics import f1_score,precision_score,recall_score,confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
class MyDecisionTree:
  def __init__(self):
    self.df = None
    self.X = None
    self.y = None
    self.X_train = None
    self.X_test = None
    self.y_train = None
    self.y_test = None
    self.model = None

  def FileRead(self):
    file_path = str(input("Enter Your file path : "))
    df = pd.read_csv(file_path)
    print(df.info())
    self.X = df.iloc[:,1:]
    # print(self.X)
    self.y = df.iloc[:,0]
    # print(self.y)

  def Splicdata(self):
    self.X_train,self.X_test,self.y_train,self.y_test = train_test_split(self.X,self.y,random_state = 42,stratify = self.y, test_size=0.1)

  def DecisionTreeModel(self):
    self.model = DecisionTreeClassifier(criterion = "entropy",class_weight="balanced",max_depth = 10,random_state = 42)
    self.model.fit(self.X_train,self.y_train)
    print(self.model.get_depth())

  def F1_score(self):
    self.y_pred = self.model.predict(self.X_test)
    self.y_pred_t = self.model.predict(self.X_train)
    self.mmetric = confusion_matrix(self.y_test,self.y_pred)
    print("This is confusion metrix : " , self.mmetric)
    self.f1_score = f1_score(self.y_train,self.y_pred_t)
    print("This is F1 score for train : " , self.f1_score)
    self.f1_score = f1_score(self.y_test,self.y_pred)
    print("This is F1 score for test : " , self.f1_score)
    # [[TN, FP],
    # [FN, TP]]



  def VisualizeTree(self):
    plt.figure(figsize=(20, 10))
    plot_tree(
        self.model,
        filled=True,
        feature_names=self.X.columns,
        class_names=[str(cls) for cls in self.y.unique()],
        rounded=True
    )
    plt.title("Decision Tree Visualization")
    plt.show()


# HeartAttack = MyDecisionTree()
# HeartAttack.FileRead()
# HeartAttack.Splicdata()
# HeartAttack.DecisionTreeModel()
# HeartAttack.F1_score()
# HeartAttack.VisualizeTree()
# print(" Still fail Model ")
```
