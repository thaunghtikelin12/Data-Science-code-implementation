```python
# list_rate_movie = np.sort(rating_df["movieId"].unique())
# column = ['movieId','Adventure', 'Comedy', 'Action', 'Drama', 'Crime', 'Children', 'Mystery', 'Documentary', 'Animation', 'Thriller', 'Horror', 'Fantasy', 'Western', 'Film-Noir', 'Romance', 'War', 'Sci-Fi', 'Musical', 'IMAX']
# final_movie_weight = pd.DataFrame(columns = column)
for i in list_rate_movie[1:100]:
  rated_user_list = rating_df[rating_df["movieId"] == i]["userId"]
  user_rated_list = rating_df[rating_df["movieId"] == i]["rating"]
  weight_of_user = df_userweight[df_userweight["UserId"].isin(rated_user_list)]
  X = np.array(weight_of_user.iloc[:,1:], dtype=np.float32)
  y = np.array(user_rated_list , dtype=np.float32)
  model = Sequential([
    Input (shape = (19,)),
    Dense (units = 1)])
  model.compile(optimizer = "adam",loss = "mean_squared_error")
  model.fit(X,y,epochs=20,verbose=0.2)
  scaler = MinMaxScaler()
  minmaxscaler = scaler.fit_transform(model.get_weights()[0])
  # print(minmaxscaler)
  moviee_weight = minmaxscaler.reshape(-1,19)
  # print(moviee_weight)
  normalizer = Normalizer(norm = 'l1')
  final_weight = normalizer.transform(moviee_weight)
  # print(final_weight)
  # print(type(final_weight))
  final_weight = np.vstack((i,final_weight.reshape(19,-1))).flatten()
  movie_weight_df.loc[len(movie_weight_df)] = final_weight
  print(f"                  {i}                          ")
movie_weight_df.to_csv('my_data.csv', index=False)

```
