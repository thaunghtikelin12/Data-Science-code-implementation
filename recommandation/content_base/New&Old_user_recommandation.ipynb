```python
def user_recommandation(userId):
  select_user_weight =  df_userweight[df_userweight["UserId"] == userId ]
  multy_uweight_mgenre = np.array(select_user_weight.iloc[:,1:]) * np.array(df_movie_file.iloc[:,3:])
  weight_result = np.sum(multy_uweight_mgenre,axis = 1)
  movie_weight_for_user = pd.DataFrame({"Movie_name": df_movie_file["title"],"weight_for_user" : weight_result})
  interest100 = np.partition(weight_result,-100,axis=None)[-100]
  print(interest100)
  final_result = movie_weight_for_user[movie_weight_for_user["weight_for_user"] >= interest100]
  return final_result
user_recommandation(3)
```
