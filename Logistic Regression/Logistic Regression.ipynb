```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
def hypothesis(X,w):
  # X = 320,3
  # w = 3 , 1
  h = X @ w
  z = sigmoid(h)
  return z

def cost_function(X,y,w):
  m = len(y)
  predi = hypothesis(X,w)
  error = (-1/m)*np.sum(y*np.log(predi)+(1-y)*np.log(1-predi))
  return error
```

```python
def gradient_descent(X,y,alpha,iter_num):
  m, n = X.shape
  w = np.ones((n, 1))
  J_his = []
  # X = 320 x 3
  # y = 320 x 1
  # w = 3 x 1
  for i in range(iter_num):
    predict = hypothesis(X,w)
    gradient = (1/m)*(X.T @ (predict-y))
    J_his.append(cost_function(X_train,y_train,w))
    w = w - alpha*gradient

  return J_his,w

J_his,w = gradient_descent(X_train,y_train,0.001,700000)
J_his[699990:]

#0.6839402255654358
#0.408410538205188
#0.2839628145287589
#0.27279721490691344
```
